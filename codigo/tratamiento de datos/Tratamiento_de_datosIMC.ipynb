# Tratamiento de datos de IMC para fusionarlos con SOLUSTDAtas_tratado.csv

```python
import pandas as pd
```

```python
df = pd.read_csv('SOLUSTDAtas_tratado.csv')
df.head()
```

```python
df_IMC = pd.read_csv('Indice_miedo_codicia.csv')
df_IMC.head()
```

```python
df_IMC['date'] = pd.to_datetime(df_IMC['date'], format='%d-%m-%Y')  # Asigno el tipo datetime a la columna date
expanded_rows = []
for index, row in df_IMC.iterrows():    # Recorro todas las filas del dataframe
    date = row['date']  # Asigno a la variable date el valor de la columna date en esta fila
    value = row['value']    # Asigno a la variable value el valor de la columna value en esta fila
    for hour in range(24):  # Creo 24 filas, una por cada hora del día y las añado a la lista
        new_date = date.replace(hour=hour)
        expanded_rows.append({'date': new_date, 'value': value})
expanded_df = pd.DataFrame(expanded_rows)   # Genero un dataframe usando la lista
```

```python
start_date = '2020-08-12 03:00:00'  # Fecha de inicio
end_date = '2024-10-09 00:00:00'    # Fecha de fin
# Selecciono las filas cuya fecha sea posterior a start_date y anterior a end_date
filtered_df = expanded_df[(expanded_df['date'] >= start_date) & (expanded_df['date'] <= end_date)]
# Ordeno el dataframe por fecha, de la más antigua a la más reciente
df_fil_sort = filtered_df.sort_values('date').copy()
df_fil_sort
```

```python
df['date'] = pd.to_datetime(df['date']) # Trasnformo la columna date al tipo datetime
df_fil_sort['date'] = pd.to_datetime(df_fil_sort['date'])   # Trasnformo la columna date al tipo datetime 
# Uno los dos dataframes a través de la fecha, agregando el valor correspondiente de la columna value cuando las fechas coincidan 
df = pd.merge(df, df_fil_sort, on='date', how='left')
df.to_csv('SolAtasIMC_tratado.csv', index=False)    # Guardo el dataframe
```
